### Create training and testing data

The data is from 7/1/15 to 5/24/16, totaling 329 days. We take the last 20 days as testing set. The other 309 days before them as the training set. When creating the training data, we use 7 days as the length of each sample when splitting the data.

```python
def split_train_test(df, lag=20):
    train = df.iloc[:,:-lag]
    test = df.iloc[:,-lag:]
    return train, test


def get_X_y(df, lag=7):        
    dat = df.values
    n, m = dat.shape
    
    tmp = np.empty([0, lag+1])
    
    for i in range(n):
        for j in range(m - lag - 1):
            x = dat[i, j:(j+lag+1)]
            tmp = np.vstack([tmp, x])
    X = tmp[:,:-1]
    y = tmp[:, -1]
    return X, y



train_df, test_df = split_train_test(df)
print(train_df.shape)
print(test_df.shape)

train_X, train_y = get_X_y(train_df)
test_X, test_y = get_X_y(test_df)
```

### Deep Neural Network Model


```python
import tensorflow as tf
import numpy as np
from tensorflow.contrib.learn import SKCompat, Estimator

def get_estimator(learning_rate=0.01):
    return SKCompat(Estimator(model_fn=_model_fn, params={'learning_rate': learning_rate}))

def _model_fn(features, labels, mode, params):
    
    print('****')
    print(features)
    h1 = tf.layers.dense(features, 7, activation=tf.nn.relu)
    h2 = tf.layers.dense(h1, 4, activation=tf.nn.relu)
    h3 = tf.layers.dense(h2, 4, activation=tf.nn.relu)
    h4 = tf.layers.dense(h3, 4, activation=tf.nn.relu)
    out = tf.layers.dense(h4, 1, activation=None)        
    
    # reshape predictions
    predictions = tf.reshape(out, [-1])
    predictions_dict = {'time': predictions}
    
    # 
    loss = tf.reduce_mean(tf.abs(labels - predictions))

    train_op = tf.contrib.layers.optimize_loss(
            loss = loss,
            global_step = tf.contrib.framework.get_global_step(),
            learning_rate = params['learning_rate'],
            optimizer = tf.train.AdamOptimizer)
    
    return predictions, loss, train_op



dnn_estimator = get_estimator()

# this runs the optimizer for one 1000 steps, using 10 observations in each step
dnn_estimator.fit(train_X, train_y, batch_size=10, steps=1000)

# now, evaluate loss on the test set
dnn_estimator.score(test_X, test_y)
```
